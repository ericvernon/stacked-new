```python
def how_sparse_is_it(dataset_id):
    X, y = load_dataset(dataset_id)
    import numpy as np

    # 1. Absolute Range (max - min) for each of the 9 features
    absolute_ranges = np.ptp(X, axis=0)
    # Equivalently: np.max(X, axis=0) - np.min(X, axis=0)
    
    # 2. Relative Range (Range / Mean)
    relative_ranges = np.ptp(X, axis=0) / np.mean(X, axis=0)
    
    # 3. Min-Max Normalized Range (Range / Max)
    # Useful if you want to see how the spread compares to the absolute highest value
    min_max_relative = np.ptp(X, axis=0) / np.max(X, axis=0)
    
    # Print a clean summary of the spread
    means = np.mean(X, axis=0)
    print(f'Dataset: {dataset_id}')
    for i in range(X.shape[1]):
        print(f"Feature {i}: Mean = {means[i]:.2f} | Range = {absolute_ranges[i]:.2f} | Relative Spread = {relative_ranges[i]:.2f}x")
```

```python
how_sparse_is_it(451)  # Breast Cancer Coimbra, 161 patterns
```

```python
how_sparse_is_it(537)  # Cervical Cancer Behavior Risk, 72 patterns
```
